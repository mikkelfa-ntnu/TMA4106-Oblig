```python
import numpy as np
import matplotlib.pyplot as plt

def square_wave_approximation(x, num_terms):
    approximation = np.zeros_like(x)
    for n in range(1, num_terms + 1, 2):
        approximation += (4 / (np.pi * n)) * np.sin(n * x)
    return approximation

def plot_approximations(num_terms_list):
    x = np.linspace(-np.pi, np.pi, 400)
    square_wave = np.sign(np.sin(x))
    
    plt.figure(figsize=(10, 6))
    
    for num_terms in num_terms_list:
        approximation = square_wave_approximation(x, num_terms)
        plt.plot(x, approximation, label=f'{num_terms} bølger')
    
    plt.plot(x, square_wave, 'k--', label='Ideel firkantbølge')
    plt.legend()
    plt.title('Firkantbølge ved fourierapproksimasjon')
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.grid(True)
    plt.savefig("fourerapproksimasjon")

num_terms_list = [1, 3, 5, 10, 20]
plot_approximations(num_terms_list)
```
